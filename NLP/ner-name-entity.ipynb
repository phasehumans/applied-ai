```python
import nltk
from nltk import ne_chunk, pos_tag, word_tokenize
from nltk.tree import Tree

# ne_chunk - named entity recognition(NER)
# tree - obj, hier struct
```

```python
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab')
```

```python
def ner_nltk(text):
  tokens= word_tokenize(text)
  # print(tokens)
  pos_tags= pos_tag(tokens)
  # print(pos_tags)
  named_entities= ne_chunk(pos_tags)
  # print(named_entities)

  entities= []
  for i in named_entities:
    if isinstance(i,Tree):
      entity= " ".join([token for token, pos in i.leaves()])
      label= i.label()
      entities.append((entity, label))
      # print(entities)
  return entities

  # flow: tokenize -> pos to tkn -> ner on pos tkn -> check for tree obj(named entity)
  # -> extract leaves(wrds) & join w/spaces -> append label, entity
```

```python
text= """Reliance Industries, led by Mukesh Ambani, announced a new partnership with Jio Platforms in
Mumbai on 15th July 2023. The project aims to invest â‚¹20,000 crores in digital infrastructure across
India"""

entities= ner_nltk(text)

print("Named Entities: ")
for entity,label in entities:
    print(f"{entity} ({label})")
```
