```python
# ===============================
# ğŸ§  Smart Chat Analyzer (Colab)
# ===============================

# Install required libraries (only the first time)
!pip install textblob nltk -q

# Import libraries
from textblob import TextBlob, download_corpora
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import nltk

# Download NLTK and TextBlob corpora
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)  # required by TextBlob
download_corpora.download_all()  # ensures TextBlob data is ready
```

```python
def analyze_chat(text):
    if not text.strip():
        return "âš ï¸ Please enter a valid text."

    print("\n===== ğŸ§  Smart Chat Text Analyzer =====")

    # 1ï¸âƒ£ Sentiment Analysis
    blob = TextBlob(text)
    polarity = blob.sentiment.polarity

    if polarity > 0:
        sentiment = "ğŸ˜Š Positive"
    elif polarity < 0:
        sentiment = "ğŸ˜ Negative"
    else:
        sentiment = "ğŸ˜ Neutral"

    print(f"\n1ï¸âƒ£ Sentiment: {sentiment} (Polarity: {polarity:.2f})")

    # 2ï¸âƒ£ Spelling Correction
    corrected_text = str(blob.correct())
    print(f"\n2ï¸âƒ£ Corrected Text: {corrected_text}")

    # 3ï¸âƒ£ Keyword Extraction
    stop_words = set(stopwords.words('english'))
    words = word_tokenize(text.lower())
    filtered = [w for w in words if w.isalnum() and w not in stop_words]
    freq = Counter(filtered)

    print("\n3ï¸âƒ£ Top 5 Keywords:")
    for word, count in freq.most_common(5):
        print(f" - {word}: {count}")

    # 4ï¸âƒ£ Simple Summary
    sentences = text.split('.')
    summary = '. '.join(sentences[:2]) + '.' if len(sentences) > 2 else text
    print(f"\n4ï¸âƒ£ Summary: {summary.strip()}")

    # 5ï¸âƒ£ Suggested Reply
    if polarity > 0:
        suggestion = "Great! You seem happy ğŸ˜„. Keep it up!"
    elif polarity < 0:
        suggestion = "I'm sorry to hear that ğŸ˜”. Stay strong!"
    else:
        suggestion = "Okay ğŸ‘. Anything else you'd like to discuss?"
    print(f"\n5ï¸âƒ£ Suggested Reply: {suggestion}")


# Example text to test
sample_text = "I am really excited about my new project! It's going great and I love coding in Python."

# Run the analyzer
analyze_chat(sample_text)
```
