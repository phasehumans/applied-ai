```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt


text = """
Sample text
Natural Language Processing (NLP) is a field of Artificial Intelligence that focuses on the interaction between computers and humans.

NLP techniques include tokenization, stemming, lemmatization, and part-of-speech tagging. Applications of NLP include sentiment analysis and
Preprocessing steps like removing stopwords, handling punctuation, and converting text to lowercase are important
in NLP pipelines. NLP plays a major role in search engines, virtual assistants, healthcare, and social media analysis.
"""


stopwords = set(STOPWORDS)


wordcloud = WordCloud(width=400, height=200,
                      background_color="orange",
                      stopwords=stopwords,
                      colormap="viridis").generate(text)


plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
